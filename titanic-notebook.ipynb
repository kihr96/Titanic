# Titanic survival prediction

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
train = pd.read_csv("train.csv")
train.head()
```

# Missing Values

```python
train.isnull().sum().sort_values(ascending=False)
```

# Focus on Cabin: let's extract more meaning: splitting the section from the number and filling NaNs

```python
train['Cabin'].fillna('XXXX', inplace=True)
train['CabSection'] = train['Cabin'].str.slice(0, 1)
train['CabNumber'] = train['Cabin'].str.slice(1, 4)
train.head()
```

# Let's reorder the attributes in a more user friendly way:

```python
train = train[['Name', 'Age', 'Sex', 'SibSp', 'Parch', 'Ticket', 'Pclass', 'Fare', 'CabSection', 'CabNumber', 'Embarked', 'Survived']]
train.head()
```
